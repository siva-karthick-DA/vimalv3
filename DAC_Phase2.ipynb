Calculate the Total vaccination on date wise

```python
z=0
total_vac1=[]
for x in data_manu["location"].unique():
     for y in data_manu["location"]:
        if x == y:
            total= 0 + data_manu["total_vaccinations"].iloc[z]
            z=z+1
     total_vac1.append(total)
```

```python
z=0
total_vac2=[]
for x in data["country"].unique():
     for y in data["country"]:
        if x == y:
            total= 0 + data["total_vaccinations"].iloc[z]
            z=z+1
     total_vac2.append(total)

```

Country wise vaccination

```python
x=data_manu["location"].unique()
y=total_vac1
plt.figure(figsize=(10,10))
plt.ylabel("Locatoion")
plt.xlabel("Total_vaccination")
plt.barh(x,y,color=["red",'green','yellow'])
plt.show()
```

Calculate  the peple_vaccinated and people_fully_vaccinated for countries

```python
z=0
people_vac=[]
for x in data["country"].unique():
     for y in data["country"]:
        if x == y:
            total= 0 + data["people_vaccinated"].iloc[z]
            z=z+1
     people_vac.append(total)
```

```python
z=0
people_fulvac=[]
for x in data["country"].unique():
     for y in data["country"]:
        if x == y:
            total= 0 + data["people_fully_vaccinated"].iloc[z]
            z=z+1
     people_fulvac.append(total)
```

Create dataframe for country wise vaccination distibution

```python
datafre={"country":data['country'].unique(),"People_Vaccinated":people_vac,"people_Fully_vaccinated":people_fulvac}
new_data=pd.DataFrame(datafre)
new_data
```

```python
x1=new_data["People_Vaccinated"]
y1=new_data["people_Fully_vaccinated"]
plt.figure(figsize=(16,4))
plt.subplot(1,4,1)
plt.scatter(x1,y1,color="red")
plt.subplot(1,4,2)
plt.plot(x1,y1,color='blue')
plt.subplot(1,4,3)
sns.kdeplot(x1,shade=True,label="people_vaccinated")
plt.legend()
plt.subplot(1,4,4)
sns.distplot(y1,color='green',label='people_fully_vaccinated')
plt.legend()
plt.show()
```

From above graph we can say that these two fields are  positively relatived
